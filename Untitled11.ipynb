```python
import pandas as pd
import math


# Load dataset

data = pd.read_csv("gold.csv")

print("Columns:", data.columns)


# Select features

X = data[['SPX', 'USO', 'SLV', 'EUR/USD']].values
Y = data['GLD'].values



# Initialize weights

w = [0.5, 0.3, 0.2, 0.1]
b = 0.1

learning_rate = 0.000001
epochs = 200



# Activation function

def sigmoid(z):
    return 1 / (1 + math.exp(-z))



# Training
for epoch in range(epochs):

    total_loss = 0

    for i in range(len(X)):

        x = X[i]
        target = Y[i]


        # normalization

        x = [
            x[0] / 1000,
            x[1] / 100,
            x[2] / 100,
            x[3] / 2
        ]
        target = target / 1000


        # forward pass

        z = (
            x[0]*w[0] +
            x[1]*w[1] +
            x[2]*w[2] +
            x[3]*w[3] +
            b
        )

        y = sigmoid(z)


        # error

        error = y - target
        dz = error * y * (1 - y)


        # update weights
        w[0] -= learning_rate * x[0] * dz
        w[1] -= learning_rate * x[1] * dz
        w[2] -= learning_rate * x[2] * dz
        w[3] -= learning_rate * x[3] * dz
        b -= learning_rate * dz

        # loss
        total_loss += 0.5 * (y - target) ** 2

    if epoch % 20 == 0:
        print("Epoch:", epoch, "Loss:", total_loss)


# Final output
print("\nFinal Weights:", w)
print("Final Bias:", b)


# Test prediction
test = X[-1]

test = [
    test[0]/1000,
    test[1]/100,
    test[2]/100,
    test[3]/2
]

z = (
    test[0]*w[0] +
    test[1]*w[1] +
    test[2]*w[2] +
    test[3]*w[3] +
    b
)

pred = sigmoid(z)

print("Prediction:", pred)
```
